Importe Numpy como np

```python
import numpy as np
```

Crie uma matriz de 10 zeros

```python
np.zeros(10)
```

Crie uma matriz de 10 ones

```python
np.ones(10)
```

Crie uma matriz de 10 cincos

```python
np.ones(10) * 5
```

Crie um array de inteiros de 10 ate 50

```python
np.arange(10, 50)
```

Crie um array dos numeros pares de 10 ate 50

```python
np.arange(10,50, 2)
```

Crie uma matriz 3x3 com valores variando de 0 ate 8

```python
np.arange(0, 9).reshape(3,3)
```

Crie uma matriz identidade 3x3

```python
np.eye(3,3)
```

Use numpy para gerar numeros aleatorios entre 0 e 1

```python
np.random.rand(10)
```

Use numpy para gerar um array de 25 numeros aleatorios tirados de uma distribuicao normal

```python
np.random.randn(25)
```

Crie a seguinte matriz:

```python
np.arange(0.01, 1.01, 0.01)
```

Crie um array de tamanho 20 igualmente espacado entre 0 e 1

```python
np.linspace(0, 1, 20)
```
