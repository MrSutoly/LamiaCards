```python
import pandas as pd
```

Leia o arquivo Salaries.csv como um DF chama "sal"

```python
sal = pd.read_csv('Salaries.csv')
```

Verifique o head do DF

```python
sal.head()
```

Use o metodo .info() pra descobrir quantas entradas existem

```python
sal.info()
```

Qual e o BasePay medio?

```python
sal['BasePay'].mean()
```

Qual e a maior quantidade de OvertimePay no conjunto de dados?

```python
sal['OvertimePay'].max()
```

Qual e o cargo de Joseph Driscoll? Nota:Use todas as maisculas, caso contrario voce pode obter uma resposta que nao coincida(ha tbm um joseph driscoll com minusculas)

```python
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

Quanto JOSEPH DRISCOLL ganha (incluindo beneficios)

```python
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
```

Qual o nome da pessoa mais bem paga(incluindo beneficios)

```python
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].max()]
```

Qual o nome da pessoa paga mais baixa (incluindo beneficios)?  voce percebe algo estranho sobre o quanto ele ou ela e paga?

```python
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].min()] #o salario esta negativo
```

Qual foi a media BasePay de todos os funcionario por ano?(2011 - 2014)

```python
group = sal.groupby('Year')
```

```python
group['BasePay'].mean()
```

Quantos titulos de trabalho unicos existem?

```python
sal['JobTitle'].nunique()
```

Qual os 5 principaos empregos mais comuns

```python
sal['JobTitle'].value_counts().head()
```

Quantos titulos de trabalho foram representados por apenas uma pessoa em 2013?(por exemplo, titulos de trabalho com apenas uma ocorrencia em 2013)

```python
sal[sal['Year'] == 2013]['JobTitle'].value_counts().loc[lambda x: x == 1].count()
```

Quantas pessoas tem a palavra chefe em seu cargo?(isso e bastante complicado)

```python
sal['JobTitle'].apply(lambda x: 'CHIEF' in x.upper()).sum()
```

Bonus: existe uma correlacao entre o comprimento da sequencia do tutlo do trabalho e salario?

```python
sal['Tamanho_String'] = sal['JobTitle'].apply(len) #esse eu tive um pouco de dificuldade ai fui para a solucao
```

```python
sal[['Tamanho_String', 'TotalPayBenefits']].corr() 
```
