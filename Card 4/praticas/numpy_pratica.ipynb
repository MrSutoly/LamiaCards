```python
arr = np.arange(1, 21) #faz um array indo de 1 a 20
```

```python
arr.reshape((4, 5)) #joga o array em forma de matriz 4x5
```

```python
arr = np.arange(1, 26).reshape(5,5) #faz um array de 1 ate 25 e joga pra uma matriz 5x5
```

```python
arr
```

```python
arr[2, 2] #pega o numero que ta na linha 2 e na coluna 2
```

```python
arr[4:][:] #pega a ultima linha da matriz
```

```python
arr[1:3, 1:3] #pega a linha 1 e 2 e as coluna 1 e 2
```

```python
arr[:,0] = 0 #define a 1 coluna como 0
```

```python
arr
```

```python
arr[arr > 20] #pega so os numeros maiores q 20 da matriz
```

```python
arr + 100 #soma 100 em todos os indices da matriz
```

```python
arr.mean() #pega a media da matriz
```

```python
arr.std() #pega o DP dos valores da matriz
```

```python
arr.sum(axis = 0) #soma as coluna da matriz
```

```python
arr.argmax() #pega o indice de onde ta o maior numero da matriz
```
