```python
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on = ['key1', 'key2']) #junta onde as keys forem iguais
```

```python
pd.merge(esquerda, direita, how = 'outer', on = ['key1', 'key2']) #junta os df e onde nao tiver numero coloca NaN
```

```python
pd.merge(esquerda, direita, how = 'right', on = ['key1', 'key2']) #deixa todas as linhas do df direita e junta com os q bater as keys com o da esquerda
```

```python
pd.merge(esquerda, direita, how = 'left', on = ['key1', 'key2']) #faz o msm com o de cima mas agr mantendo o da esquerda
```
