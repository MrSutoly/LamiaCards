```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()


df = pd.read_csv("taylor_swift_spotify.csv") #lendo um csv baixado da internet
df
```

```python
plt.figure(figsize=(10,6)) #define tamanho do grafico
sns.histplot(df["danceability"], kde=True) #histograma o kde true add a curva de densidade 
plt.title("DistribuiÃ§Ã£o de Danceability") #titulo pro grafico
plt.show()
```

```python
plt.figure(figsize=(10,6)) #tamanho do grafico novo
sns.scatterplot(x="energy", y="popularity", data=df) #grafico de dispersao cada ponto representa uma musica
plt.title("Energia vs Popularidade")
plt.show()
```

```python
df["year"] = pd.to_datetime(df["release_date"]).dt.year #converte a coluna realease_data pra formato de data o .dt.year pega so o ano

df["decade"] = (df["year"] // 10) * 10 #divide por 10 e multiplica por 10 pra formar decadas, tipo, 202 vira 2020

plt.figure(figsize=(12,6)) #define tamanho
sns.boxplot(x="decade", y="energy", data=df) #grafico em boxplot
plt.title("Energy por Era")
plt.show()
```

```python
sns.pairplot(df[["danceability","energy","popularity","valence"]]) #cria varios graficos
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="tempo", y="danceability", data=df) #grafico em dispersao onde cad ponto e uma musica
plt.title("RelaÃ§Ã£o entre Tempo e Danceability")
plt.show()
```

```python
top10 = df.sort_values("popularity", ascending=False).head(10) #cria um top 10 das mais populares
sns.barplot(x="popularity", y="name", data=top10) #grafico em barra
plt.title("Top 10 mÃºsicas Taylor Swift")
plt.show()
```
