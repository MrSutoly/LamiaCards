Standard Deviation and Variance (Exercicio)

```python
import numpy as np 
import matplotlib.pyplot as plt #importa as bibliotecas
```

```python
incomes = np.random.normal(100.0, 20.0, 10000) #10 mil numeros de media 100 e DP 20
```

```python
plt.hist(incomes, 50) #cria um histograma
```

```python
incomes.std() #pega o desvio padrao ou DP
```

```python
incomes.var() #pega a variancia
```

O exercicio pede para alterar os parametros e ver como isso afeta o corpo da distribuicao

```python
incomes = np.random.normal(1003.0, 50.0, 10000) #muda os paraemtros de media e DP
```

```python
plt.hist(incomes, 50)
```
