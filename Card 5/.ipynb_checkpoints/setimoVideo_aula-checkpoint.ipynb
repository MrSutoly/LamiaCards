```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vals = np.random.normal(0, 0.5, 10000) #via distribuicao normal cria 10000 valores com dp de 0.5 e media 0
```

```python
plt.hist(vals, 50) #faz um histograma com os valores
plt.show()
```

```python
np.percentile(vals, 50) #calcula o percentil 50 dos dados ou seja 50% dos valores estao acima e abaixo desse numero
```

```python
np.percentile(vals, 90) #calcula o percentil 90, ou seja 10 estao acima dele e 90 abaixo
```

```python
np.percentile(vals, 20) #o mesmo so que com o percentil de 20
```

Moments: Mean, Variance, Skew, Kurtosis

```python
vals = np.random.normal(0, 0.5, 10000) #cria uma distribuicao normal igual a de cima
```

```python
plt.hist(vals, 50)
plt.show() #cria o histograma e mostra o grafico
```

```python
np.mean(vals) #pega a media dos valores
```

```python
np.var(vals) #pega a variancia dos valores
```

```python
import scipy.stats as sp
```

```python
sp.skew(vals) #pega o skew dos valores
```

```python
sp.kurtosis(vals) #pega o kurtosis dos valores
```
