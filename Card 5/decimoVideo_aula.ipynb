Covariancia e Correlecao

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x] #calcula a media de x

def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)  #calcula a covariancia de x e y

pageSpeeds = np.random.normal(3.0, 1.0, 1000) #faz 1000 valores aleatorios com media 3 e dp 1
purchaseAmount = np.random.normal(50.0, 10.0, 1000) #faz 1000 com media 50 e dp 10

plt.scatter(pageSpeeds, purchaseAmount) #grafico de dispersao com os dados acima

covariance(pageSpeeds, purchaseAmount) #usa a funcao criada acima pra calcular a covariancia
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds #randomiza valores igual la em cima mas agr depende da velocidade da pagina

plt.scatter(pageSpeeds, purchaseAmount) #grafico de dispersao

covariance(pageSpeeds, purchaseAmount) #covariancia com os dados acima
```

```python
def correlation(x, y):
    stddevx = x.std() #pega o DP de x
    stddevy = y.std() #pega o DP de y
    return covariance(x, y) / stddevx / stddevy #calcula a correlacao entre x e y

correlation(pageSpeeds, purchaseAmount) #chama a funcao
```

```python
np.corrcoef(pageSpeeds, purchaseAmount) #calcula a correlacao com a funcao do numpy
```

```python
purchaseAmount = 100 - pageSpeeds * 3 #valores de compra diminuem em relacao a velocidade da pagina que aumenta
plt.scatter(pageSpeeds, purchaseAmount) #grafico pra mostrar q quanto maior a velocidade da pagina menor a compra
correlation(pageSpeeds, purchaseAmount) #correlcao dos dados acima
```
