Mean, Median, Mode, and introducing NumPy

```python
import numpy as np
```

```python
incomes = np.random.normal(27000, 15000, 10000) #gera 100 valores com dp de 1500 e media de 2700
```

```python
incomes
```

```python
np.mean(incomes) #pega a media dos valores
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.hist(incomes, 50) #faz um histograma com 50 barrinhas
```

```python
np.median(incomes) #pega a mediana dos valores
```

```python
incomes = np.append(incomes, 1000000000)
```

```python
np.median(incomes) #pega a mediana dos valores
```

```python
np.mean(incomes) #pega a media dos valores
```
