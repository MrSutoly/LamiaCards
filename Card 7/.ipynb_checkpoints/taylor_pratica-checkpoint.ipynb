```python
import pandas as pd
import matplotlib.pyplot as plt #importa as bibliotecas
```

```python
df = pd.read_csv('/kaggle/input/taylor-swift-spotify-dataset/taylor_swift_spotify.csv') #le o csv
```

```python
df['album'].value_counts().plot(kind = 'barh') #pega cada album album e faz o grafico
plt.title('Número de músicas por Álbum') #titulo do grafico
plt.show()
```

```python
albuns = ["Fearless (Taylor's Version)", "Red", "1989"] #uma lista so com esses 3 albuns
df_fr = df[df['album'].isin(albuns)] #cria um df novo com somente os albuns da lista
```

```python
top10 = df_fr.sort_values('popularity', ascending = False).head(10) #vai do maior pro menor em questao de popularidade

plt.figure(figsize = (10,5))
plt.bar(top10['name'], top10['popularity'], color = 'brown')
plt.xticks(rotation = 45, ha = 'right')
plt.title('Top 10 músicas mais populares (Fearless, Red, 1989)')
plt.ylabel('Popularidade') #cria o grafico e define textos
plt.show()
```

```python
media_pop = df_fr.groupby('album')['popularity'].mean() #pega a media de popularidade de cada album

plt.bar(media_pop.index, media_pop.values, color = ['deepskyblue', 'gold', 'red'])
plt.title('Popularidade média por álbum')
plt.ylabel('Popularidade média') #faz o grafico com os dados manipulado pela primeira linha
plt.show()
```

```python
media_dance = df_fr.groupby('album')['danceability'].mean() #pega a media de danceabilidade de cada album

plt.bar(media_dance.index, media_dance.values, color = ['deepskyblue', 'gold', 'red'])
plt.title('Danceability média por álbum')
plt.ylabel('Danceability') #faz o grafico
plt.show()
```

```python
media_valence = df_fr.groupby('album')['valence'].mean() #media de felicidade de cada album

plt.bar(media_valence.index, media_valence.values, color = ['deepskyblue', 'gold', 'red'])
plt.title('Felicidade Média')
plt.ylabel('Valence') #faz o grafico
plt.show()
```

```python
red = df_fr[df_fr['album'] == 'Red'].sort_values(by='popularity', ascending = False) #faz um df so com o album RED do maior pro menor

plt.figure(figsize = (12,5))
plt.bar(red['name'], red['popularity'], color = 'red')
plt.xticks(rotation = 75, ha = 'right')
plt.title('Popularidade das músicas do álbum RED')
plt.ylabel('Popularidade') #faz o grafico da popularidade de cada musica do RED
plt.show()
```

```python
a1989 = df_fr[df_fr['album'] == '1989'].sort_values(by='popularity', ascending = False) #df so com os dados do 1989 do maior pro menor

plt.figure(figsize = (12,5))
plt.bar(a1989['name'], a1989['popularity'], color = 'deepskyblue')
plt.xticks(rotation = 75, ha = 'right')
plt.title('Popularidade das músicas do álbum 1989')
plt.ylabel('Popularidade') #faz o grafico de popularidade de cada musica do 1989
plt.show()
```

```python
fearless = df_fr[df_fr['album'] == "Fearless (Taylor's Version)"].sort_values(by='popularity', ascending = False) #df so com os dados do Fearless
#do maior pro menor

plt.figure(figsize = (12,5))
plt.bar(fearless['name'], fearless['popularity'], color = 'gold')
plt.xticks(rotation = 75, ha = 'right')
plt.title('Popularidade das músicas do álbum Fearless')
plt.ylabel('Popularidade') #cria o grafico de popularidade de cada musica do fearless
plt.show() 
```
